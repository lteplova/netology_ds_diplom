```python
trials = 1000000 # Количество попыток
cost = trials*1  # Стоимость всех игр
successes = 0    # Количество успехов
balance = 0      # Счет игрока
cost_win = 0     # Стоимость выигрыша

for i in range(trials):
    res = np.random.randint(0, 1000)
    balance -=1
    if res == 777:
        cost_win +=200
        successes +=1
    elif res == 999:
        cost_win +=100
        successes +=1
    elif res == 555:
        cost_win +=50
        successes +=1
    elif res == 333:
        cost_win +=15
        successes +=1
    elif res == 111:
        cost_win +=10
        successes +=1
    elif res%10 == 77:
        cost_win +=5
        successes +=1
    elif res%10 == 7:
        cost_win +=3
        successes +=1
    elif res%100 == 0:
        cost_win +=2
        successes +=1
    elif res%10 == 0:
        cost_win +=1
        successes +=1

p_monte_carlo = successes / trials    # Частота моделирования
p = (5 + 20 + 200) / 1000             # Теоретическая вероятность любого выигрыша
lost = cost - abs(cost_win)           # Сколько денег было проиграно
game = cost_win/cost                 # Средний выигрыш за одну игру

print('Частота:', p_monte_carlo)
print('Теоретическая вероятность наступления выигрыша:', p)
print('Счет игрока:', balance + cost_win)
print('Стоимость всех выигрышей:', cost_win)
print(f'Количество выйгрышей из {trials} попыток: {successes}')
print('Проиграно:', lost)
print(f'Выгода одной игры: {game} рублей' )
```
