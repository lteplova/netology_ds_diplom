```python
# перевод имен колонок датафрейма в список
all_colums = df.columns.tolist()

# словарь с нужными для работы колонками
columns_names_dict = {
                1 : 'surgery?',
                2 : 'Age',
                4 : 'rectal temperature',
                5 : 'pulse',
                6 : 'respiratory rate',
                7 : 'temperature of extremities',
                11 : 'pain',
                23 : 'outcome'
}

# удаление ненужных колонок путем сравнения списка всех колонок со словарем искомых
for k in columns_names_dict:
    for el in all_colums:
        if el == k-1:
            all_colums.remove(el)
df_edited = df.drop(df.columns[all_colums], axis = 1)

# переименование колонок
df_edited.columns = [ col for col in columns_names_dict.values() ]

df_edited.head(5)
```

## Выбросы и удаление


У здоровой взрослой лошади температура тела колеблется в пределах
   

*  37,5–38,5 градусов,
*  пульс – 24–42 удара в минуту,
*  дыхание – 8–16 вдохов и выдохов в минуту.



У жеребят эти параметры несколько выше:
* температура до 39 градусов,
* пульс новорожденных может достигать 130 ударов в минуту,
* а число дыханий доходить до 85

Через полгода эти цифры меняются:
* 50–72 удара и
* 14–15 вдохов и выдохов в минуту.

При изменении температуры, пульса и дыхания надо учитывать окружающую обстановку, температуру среды и состояние лошади. Во время работы и после нее, в новой обстановке, при половом возбуждении, на последних месяцах жеребости, в жару в организме скорость обмена веществ возрастает, а следовательно, и указанные параметры тоже.
Например, во время скачки температура тела может подняться на три градуса выше нормы.
В холод, в периоды отдыха и сна все процессы в организме замедлены, и показатели – на минимальном уровне нормы.
