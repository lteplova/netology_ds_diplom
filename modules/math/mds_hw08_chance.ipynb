Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

```python
import scipy.stats as sps
from scipy.stats import t
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
grid = np.linspace(-7, 7, 1000)  # сетка для построения графика
loc_values = [1, 1.5, 1.7, 2, 1]  # набор значений параметра loc
scale_values = [0.5, 1, 1.2, 1.5, 2]  # набор значений параметра scale

plt.figure(figsize=(22, 6))

for i, (a, scale) in enumerate(zip(loc_values, scale_values)):
    plt.plot(grid, sps.norm(a, scale).pdf(grid), lw=5,
             label='$\mathcal{N}' + '({}, {})$'.format(a, scale))

plt.legend(fontsize=16)
plt.title('Плотности нормального распределения', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```

```python
grid = np.linspace(-7, 7, 1000)  # сетка для построения графика
loc_values = [1, 1.5, 1.7, 2, 1]  # набор значений параметра loc
scale_values = [0.5, 1, 1.2, 1.5, 2]  # набор значений параметра scale

plt.figure(figsize=(22, 6))

for i, (a, scale) in enumerate(zip(loc_values, scale_values)):
    plt.plot(grid, sps.expon(a, scale).pdf(grid), lw=5,
             label='$\mathcal{N}' + '({}, {})$'.format(a, scale))

plt.legend(fontsize=16)
plt.title('Плотности экспоненциального распределения', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```

```python
grid = np.linspace(-7, 7, 1000)  # сетка для построения графика
loc_values = [1, 1.5, 1.7, 2, 1]  # набор значений параметра a
scale_values = [0.2, 1, 1.2, 2, 2.5]  # набор значений параметра sigma (дисперсия)

plt.figure(figsize=(22, 6))

for i, (a, scale) in enumerate(zip(loc_values, scale_values)):
    plt.plot(grid, sps.t(a, scale).pdf(grid), lw=5,
             label='$\mathcal{N}' + '({}, {})$'.format(a, scale))

plt.legend(fontsize=16)
plt.title('Плотности распределения Стьюдента', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```
