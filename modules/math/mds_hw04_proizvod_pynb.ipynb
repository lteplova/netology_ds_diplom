Посчитайте значение первой и второй производной функции:

В точке x = 1.

```python
# значение первой производной функции
def f(x):
    return x**5 + (4 * np.sin(2*x)) + (np.cos(3*x +3))
derivative(f, 1.0, n=1, dx=1e-6)
```

```python
# значение  второй производной функции
derivative(f, 1.0, n=2, dx=1e-6)
```

```python
x = Symbol('x')
```

```python
dx_1= diff( x**5 + (4 * sin(2*x)) + (cos(3*x +3))).evalf() ## первая производная
dx_1
```

```python
res_1 = dx_1.subs(x, 1) # вычисление результата первой производной в точке x=1
```

```python
res_1
```

```python
dx_2 = diff(dx_1)
dx_2
```

```python
dx_2.subs(x, 1) # вычисление результата второй производной в точке x=1
```

Постройте график первой и второй производной функции:


На отрезке [-5; 5].

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
# построение графика первой производной функции
for el in x:
    def f(x):
        return (np.sin(2*x + 1))**5
y1 = derivative(f, x, n=1, dx=1e-6)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График первой производной')
plt.show()
```

```python
# построение графика второй производной функции
for el in x:
    def f(x):
        return (np.sin(2*x + 1))**5
y2 = derivative(f, x, n=2, dx=1e-6)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График второй производной')
plt.show()
```

```python
# построение графика первой и второй производной на одном графике
for el in x:
    def f(x):
        return (np.sin(2*x + 1))**5
y1 = derivative(f, x, n=1, dx=1e-6)
plt.plot(x, y1, label = 'dx1')
plt.xlabel('x')
plt.ylabel('y')

y2 = derivative(f, x, n=2, dx=1e-6)
plt.plot(x, y2, label = 'dx2')

plt.legend()
plt.title('График первой и второй производной')
plt.show()
```
