```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

# Обязательная часть

```python
def get_habr_search(keywords):
    URL = 'https://habr.com/ru/search/'
    params = {
            'q': keywords,
            'target_type': 'posts',
            'order': 'relevance',

            }
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:99.0) Gecko/20100101 Firefox/99.0'
            }
    req = requests.get(URL, params=params, headers=headers)
    soup = BeautifulSoup(req.text)
    search_result = soup.find_all('article', class_='tm-articles-list__item')
    print(len(search_result))
    habr_blog = pd.DataFrame()
    # print(search_result)
    for el in search_result:

        title = el.find('h2', 'tm-article-snippet__title tm-article-snippet__title_h2').text
        link = el.find('a', 'tm-article-snippet__title-link').get('href')
        date = el.find('span', 'tm-article-snippet__datetime-published').text
        row = {'date': date, 'title': title, 'link': 'https://habr.com'+link}
        habr_blog = pd.concat([habr_blog, pd.DataFrame([row])]).reset_index(drop=True)
        # print(title)

    return habr_blog

habr_blog_common = pd.DataFrame()
habr_blog_common = pd.concat([habr_blog_common, get_habr_search('python, анализ данных')]).reset_index(drop=True)
habr_blog_common = pd.concat([habr_blog_common, get_habr_search('python')]).reset_index(drop=True)

habr_blog_common.drop_duplicates()
```

# Дополнительная часть (необязательная)

```python
def get_habr_search(keywords, pages):
    habr_blog = pd.DataFrame()
    for page in range(1, pages+1):
        URL = 'https://habr.com/ru/search/page' + str(page) + '/'
        params = {
                'q': keywords,
                'target_type': 'posts',
                'order': 'relevance',
                }
        headers = {
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:99.0) Gecko/20100101 Firefox/99.0'
                }
        req = requests.get(URL, params=params, headers=headers)
        soup = BeautifulSoup(req.text)
        search_result = soup.find_all('article', class_='tm-articles-list__item')

        for el in search_result:
            title = el.find('h2', 'tm-article-snippet__title tm-article-snippet__title_h2').text
            link = el.find('a', 'tm-article-snippet__title-link').get('href')
            date = el.find('span', 'tm-article-snippet__datetime-published').text
            row = {'date': date, 'title': title, 'link': 'https://habr.com'+link}
            habr_blog = pd.concat([habr_blog, pd.DataFrame([row])]).reset_index(drop=True)

    return habr_blog

habr_blog_common = pd.DataFrame()
habr_blog_common = pd.concat([habr_blog_common, get_habr_search('python, анализ данных', 5)]).reset_index(drop=True)
habr_blog_common = pd.concat([habr_blog_common, get_habr_search('python', 5)]).reset_index(drop=True)

habr_blog_common.drop_duplicates()
```
