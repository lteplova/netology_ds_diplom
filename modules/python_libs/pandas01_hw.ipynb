# Задача 1

Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
data_movies = pd.read_csv('/content/drive/MyDrive/pandas_data/movies.csv')
data_movies.head()
```

```python
data_ratings= pd.read_csv('/content/drive/MyDrive/pandas_data/ratings.csv')
data_ratings.head()
```

```python
data_joined = data_ratings.merge(data_movies, on='movieId', how='left')

data_res = data_joined[data_joined ['rating'] == 5.0]\
                                                .filter(items = ['title', 'rating']) \
                                                .groupby('title') \
                                                .count().sort_values(by = 'rating', ascending=False)
data_res.head(10)
```

# Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df = pd.read_csv('/content/drive/MyDrive/pandas_data/power.csv')
df.head()
```

```python
df_filtered = df[((df['country'] == 'Latvia') | (df['country'] =='Estonia') | (df['country'] =='Lithuania'))
                & (df['quantity'] > 0)
                & ((df['category'] == 4) | (df['category'] == 12) | (df['category'] == 21))
                & ((df['year'] > 2005) & (df['year'] < 2010))]

df_filtered.head()
sum_quantity = df_filtered['quantity'].sum()
print(sum_quantity)
```
