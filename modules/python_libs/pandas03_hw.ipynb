# Задание 1

```python
df = pd.read_csv('/content/drive/MyDrive/pandas_data/03/visit_log.csv', sep = ';')
df.head()
```

```python
df.loc[df.traffic_source.isin(['yandex', 'google']), 'traffic_type'] = 'organic'
df
```

```python
def source(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        row['source_type'] = 'organic'
    elif row['traffic_source'] == 'paid' or row['traffic_source'] == 'email':
        if row['region'] == 'Russia':
            row['source_type'] = 'ad'
        else:
            row['source_type'] = 'other'
    else:
        row['source_type'] = row['traffic_source']
    return row

```

```python
df = df.apply(source, axis = 1)
df.head(20)
```

```python
df_rating = pd.read_csv('/content/drive/MyDrive/pandas_data/03/ratings.csv')
df = df_rating.loc[:, ['userId', 'timestamp']].groupby('userId')
df.head()
```

```python
def df_rating_count(row):
    result = row['userId'].count()
    row['count'] = result
    if result > 100:
        row['diff'] = row['timestamp'].max() - row['timestamp'].min()
        return row
```

```python
df_res = df.apply(df_rating_count)
df_result = df_res.loc[:, ['diff']]
df_result.head()

```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```
