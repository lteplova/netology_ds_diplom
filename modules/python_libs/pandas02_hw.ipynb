```python
import pandas as pd
```

```python
df_movies = pd.read_csv('/content/drive/MyDrive/pandas_data/02/movies.csv')
df_ratings = pd.read_csv('/content/drive/MyDrive/pandas_data/02/ratings.csv')
df = df_ratings.merge(df_movies, on='movieId', how='left')
df.head()
```

```python
def rating_class(row):
    if row['rating'] <= 2:
        row['class'] = 'низкий рейтинг'
    elif row['rating'] <= 4 and row['rating'] > 2:
        row['class'] = 'средний рейтинг'
    else:
        row['class'] = 'высокий рейтинг'
    return row
```

```python
df = df.apply(rating_class, axis=1)
df.head(15)
```

# Задание 2

```python

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

df_keywords = pd.read_csv('/content/drive/MyDrive/pandas_data/02/keywords.csv')
df_keywords.head()
```

```python
def region(row):

    keywords = row['keyword'].split(' ')

    for region, cities in geo_data.items():

        for city in cities:
            if city in keywords:
                row['region'] = region
    return row
```

```python
df = df_keywords.apply(region, axis = 1).fillna('undefined')
df.head(5)
```

```python
df_filter = df[(df['region'] == 'Центр') | (df['region'] == 'Дальний Восток') | (df['region'] == 'Северо-Запад')]
df_filter.head()
```

# Задание 3

```python
df_movies = pd.read_csv('/content/drive/MyDrive/pandas_data/01/movies.csv')
df_ratings = pd.read_csv('/content/drive/MyDrive/pandas_data/01/ratings.csv')
df_movies.tail()
```

```python
years = []
for year in range(1950, 2011):
    years.append(str(year))

def production_year(row):
    title = row['title'].replace('(', '').replace(')', '').split()[-1]
    # row['year'] = title
    for el in years:
        if title in years:
            row['year'] = title
        else:
            row['year'] = 1900

    return row
```

```python
df = df_movies.apply(production_year, axis = 1)
df.tail()
```

```python
data_joined = df_ratings.merge(df, on='movieId', how='left')
data_joined.tail()
```

```python
data_res = data_joined.groupby('year').mean().rating.sort_values(ascending=False)
data_res.head(100)
```
