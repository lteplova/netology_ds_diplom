### Подготовка данных

```python
#  ищем id пользователя, поставившего больше всего оценок
df_ratings.filter(items = ['user_id', 'rating']).groupby('user_id').count().sort_values(by = 'rating', ascending=False)
```

```python
# оставим в датафрейме ratings только те фильмы, который оценил  пользователь с ID 405 (он поставил больше всего оценок)
df_rating_user = df_ratings[df_ratings['user_id'] == 405].filter(items = ['user_id', 'movie_id'])
df_rating_user
```

```python
# добавляем к датафрейму df_ratings столбцы с жанарми и столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей
```

```python
# объединяем датафреймы с рейтингом и фильмами
df_joined = df_ratings.merge(df_movies, on='movie_id', how='left')
df_joined.info()
```

```python
# список колонок нового датафрейма
cols = df_joined.columns
cols
```

```python
# отфильтровываем нужные столбцы
df_joined = df_joined .filter(items = ['user_id', 'movie_id', 'rating',\
       'release_date', 'unknown', 'Action',\
       'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary',\
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',\
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'])

df_joined.head()
```

```python
# добавляем стобцы с количеством оценок и с сумарной оценкой
df = df_joined.copy()
df['sum_of_ratings'] = df['movie_id'].map(df.groupby('movie_id')['rating'].sum())
df['amount_of_ratings'] = df['movie_id'].map(df.groupby('movie_id')['rating'].count())

df.sort_values(by = 'movie_id')
```

```python
# получаем результирующий датафрейм для пользователя, который поставил больше всех оценок
df_405 = df.loc[df['user_id'] == 405]
df_405
```

```python
# преобразование даты релиза в год
df_405.release_date = df_405.release_date.dt.year
```

```python
# переименование столбца release_date в year
df_405.rename(columns={'release_date' : 'year'}, inplace = 'True')
df_405
```

### Модель и обучение

```python
# Сформировать X и y из df, выбрав нужные колонки

X, y = df_405[['year', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'sum_of_ratings', 'amount_of_ratings']], df_405['rating']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
mean_squared_error(y_test, y_pred, squared=False)
```
