```python
df = spark.read.csv('/content/drive/MyDrive/data_colab/iris.CSV', inferSchema=True, header=True)
```

```python
df.show(5)
```

```python
df.describe().show()
```

```python
df.groupBy('variety_num').count().show()
```

```python
df.columns
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import DecisionTreeClassifier
```

```python
pipeline = Pipeline(stages =
[
  VectorAssembler(inputCols=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], outputCol='Features')
]
)
```

```python
pipelineTrained = pipeline.fit(df)
```

```python
pipelineTrained.transform(df).show()
```

```python
df_features = pipelineTrained.transform(df)
```

```python
df_features.show()
```

```python
train, test = df_features.randomSplit([0.8, 0.2], seed=12345)
```

```python
train.show()
```

```python
# логистическая регрессия
lr = LogisticRegression(featuresCol = 'Features', labelCol = 'variety_num')
lrModel = lr.fit(train)
```

```python
train_res = lrModel.transform(train)
test_res = lrModel.transform(test)
```

```python
train_res.show()
```

```python
ev = MulticlassClassificationEvaluator(labelCol='variety_num')
```

```python
ev.evaluate(train_res)
```

```python
ev.evaluate(test_res)
```

```python
# модель дерева решений

dt = DecisionTreeClassifier(labelCol='variety_num', featuresCol='Features')

model = dt.fit(train)
predictions = model.transform(test)
train_result = model.transform(train)
```

```python
train_result.show()
```

```python
ev.evaluate(train_result)
```

```python
ev.evaluate(predictions)
```
